### pdf to text

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
import re
```

```python
def convert_pdf_to_txt(dir_list, file_fname, file_name):
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'cp949'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(file_fname,'rb')
    PDFPage.get_pages(fp,check_extractable=False)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0 
    caching = True
    pagenos = set()
    
    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=False):
        interpreter.process_page(page)
             
    text = retstr.getvalue()
    
    fp.close()
    device.close()
    retstr.close()
    
    #v = convert_pdf_to_txt(text)
    f = open('./data/Research_Text/{dir_list}/{file_name}.txt'.format(dir_list=dir_list,file_name=file_name),'w', -1, "utf-8")
    f.write(text)
    f.close()
    
    #수정 
    with open('./data/Research_Text/{dir_list}/{file_name}.txt'.format(dir_list=dir_list,file_name=file_name),'r', -1, "utf-8") as f1:
        with open('./data/ReWriteResearchText/{dir_list}/{file_name}.txt'.format(dir_list=dir_list,file_name=file_name), 'w', encoding='utf8') as f2:
            for line in f1:
                line = f1.readline()
                line = re.sub('^[0-9]*[.][0-9]*$|^[-]?[0-9]+[\s]?$|^[-][0-9]+[\s]?$|^[0-9]+[\s]?$|^[\s]?[-]?[0-9]+[\s]?[%][\s]?$|^[a-zA-Z]+[\s]?$|^[\s]?[-]?[0-9]+[\.][0-9]+[\s]?[%][\s]?$|^[0-9]+[\.][0-9]+[\s]?$|^[0-9]+[-][0-9]+[\s]?$|^[0-9]+\'s$|^[\s]?[-=+,#/\?:^$.@*\"※~&%ㆍ•!』\\‘|\(\)\[\]\<\>`\'…》]|^\{^\}','',line)
                f2.write(line + '\n')
    
    
  
    
    
    #return text

    
```

#### 제거해야할 것(연습)

```python
# 1. 특수문자 제거 
# 2. 그림 및 그래프 수치 제거 
# 수치는 숫자, % 
# 표 같은거 잘
# 한 줄에 한 글자 ~ 두세글자 삭제
```

```python
def re_text(file_name):
    f = open('./data/Research_Text/{file_name}.txt'.format(file_name=file_name),'a', -1, "utf-8")
    while True:
        line = f.readline()
        #그림 속 수치 및 등 제거 
        line = re.sub('^[0-9]+$|^[-]?[0-9]+[%]$|^[a-zA-Z]+$|^[-]?[0-9]+[.][0-9]+[%]$','',line)
        line = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ•!』\\‘|\(\)\[\]\<\>`\'…》]', '', line)
        if not line: break
        
    f.close()
        
```
